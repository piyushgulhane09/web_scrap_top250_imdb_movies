```python
import requests
from bs4 import BeautifulSoup as bs
from urllib.request import urlopen
import logging
```

```python
raw_page1_url = "https://www.imdb.com/chart/top/"
```

```python
page1_url_opened = urlopen(raw_page1_url)
```

```python
page1_url_html = bs(page1_url_opened.read(),'html.parser')
```

```python
main_box = page1_url_html.findAll('div',{'class':'article'})
```

```python
list_titles = main_box[0].span.div.div.findAll('td',{'class':'titleColumn'})
```

```python
list_titles[0].a.text  
```

```python
list_ratings = main_box[0].span.div.div.findAll('td',{'class':'ratingColumn imdbRating'})
```

```python
list_ratings[0].strong.text
```
